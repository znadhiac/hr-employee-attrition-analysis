# **HR ANALYTICS EMPLOYEE ATTRITION AND PERFORMANCE <br> DATA CLEANING**

## **1. Load Data**

```python
# Import Libraries
import pandas as pd
import numpy as np

# Load Data
employee_raw = pd.read_csv("Employee.csv")
education_raw = pd.read_csv("EducationLevel.csv")
performance_raw = pd.read_csv("PerformanceRating.csv")
rating_raw = pd.read_csv("RatingLevel.csv")
satisfaction_raw = pd.read_csv("SatisfiedLevel.csv")

employee = employee_raw.copy()
education = education_raw.copy()
performance = performance_raw.copy()
rating = rating_raw.copy()
satisfaction = satisfaction_raw.copy()
```

## **2. Initial Data Inspection**

```python
# INITIAL DATA INSPECTION

datasets = {
    "Employee": employee,
    "EducationLevel": education,
    "PerformanceRating": performance,
    "RatingLevel": rating,
    "SatisfiedLevel": satisfaction
}

for name, df in datasets.items():
    print(f"\n{'='*40}")
    print(f"{name} Table")
    print(f"{'='*40}")
    print("Shape:", df.shape)
    print("\nInfo:")
    df.info()
    print("\nPreview:")
    display(df.head())
```

## **3. Converting Incorrect Data Type**

```python
# CONVERT INCORRECT OBJECT TYPE INTO DATE TYPE

employee["HireDate"] = pd.to_datetime(
    employee["HireDate"], errors="coerce"
)
performance["ReviewDate"] = pd.to_datetime(
    performance["ReviewDate"], errors="coerce"
)
```

## **4. Handling Missing Values**

```python
# DOUBLE CHECK MISSING VALUES FOR ALL TABLES

datasets = {
    "Employee": employee,
    "EducationLevel": education,
    "PerformanceRating": performance,
    "RatingLevel": rating,
    "SatisfiedLevel": satisfaction
}

for name, df in datasets.items():
    print(f"\n{name} - Missing Values by Column:")
    print(df.isnull().sum())
```

No missing values were found across all datasets. Therefore, no imputation or removal was required.

## **5. Handling Duplicate Data**

```python
# CHECK DUPLICATE DATA FOR ALL TABLES

datasets = {
    "Employee": employee,
    "EducationLevel": education,
    "PerformanceRating": performance,
    "RatingLevel": rating,
    "SatisfiedLevel": satisfaction
}

for name, df in datasets.items():
    duplicates = df.duplicated().sum()
    print(f"{name} - Number of duplicate rows: {duplicates}")
```

No duplicate data were found across all datasets. Therefore, no action was required.

## **6. Handling Inconsistent Data**

```python
# CHECK CATEGORICAL VALUE CONSISTENCY FOR ALL TABLES

datasets = {
    "Employee": employee,
    "EducationLevel": education,
    "PerformanceRating": performance,
    "RatingLevel": rating,
    "SatisfiedLevel": satisfaction
}

print("=== CATEGORICAL VALUE CONSISTENCY CHECK ===")
for name, df in datasets.items():
    cat_cols = df.select_dtypes(include="object").columns
    if len(cat_cols) == 0:
        print(f"\n{name}: No categorical columns")
        continue

    print(f"\n{name}:")
    for col in cat_cols:
        unique_values = df[col].str.strip().unique()
        print(f"  {col}: {unique_values}")
```

```python
# ===============================
# CHECK FOREIGN KEYS
# ===============================

# 1. EDUCATION (Employee → EducationLevel)
print("\n=== FOREIGN KEY CHECK: EDUCATION ===")

invalid_education = (
    ~employee["Education"]
    .isin(education["EducationLevelID"])
).sum()

print(f"Invalid Education references: {invalid_education}")


# 2. EMPLOYEE ID (PerformanceRating → Employee)
print("\n=== FOREIGN KEY CHECK: EMPLOYEE ID ===")

invalid_employee_id = (
    ~performance["EmployeeID"]
    .isin(employee["EmployeeID"])
).sum()

print(f"Invalid EmployeeID references: {invalid_employee_id}")


# 3. SATISFACTION (PerformanceRating → SatisfiedLevel)
print("\n=== FOREIGN KEY CHECK: SATISFACTION ===")

satisfaction_columns = [
    "EnvironmentSatisfaction",
    "JobSatisfaction",
    "RelationshipSatisfaction"
]

for col in satisfaction_columns:
    invalid_count = (
        ~performance[col]
        .isin(satisfaction["SatisfactionID"])
    ).sum()
    print(f"Invalid {col} references: {invalid_count}")


# 4. PERFORMANCE RATING (PerformanceRating → RatingLevel)
print("\n=== FOREIGN KEY CHECK: PERFORMANCE RATING ===")

rating_columns = [
    "SelfRating",
    "ManagerRating"
]

for col in rating_columns:
    invalid_count = (
        ~performance[col]
        .isin(rating["RatingID"])
    ).sum()
    print(f"Invalid {col} references: {invalid_count}")
```

```python
# CHECK LOGICAL RANGE (EMPLOYEE TABLE)

print("\n=== LOGICAL RANGE CHECK (EMPLOYEE TABLE) ===")

# Age should be between 18 and 65
invalid_age = (~employee["Age"].between(18, 65)).sum()
print(f"Invalid Age (outside 18–65): {invalid_age}")

# Salary should be positive
invalid_salary = (employee["Salary"] <= 0).sum()
print(f"Invalid Salary (<= 0): {invalid_salary}")

# Years at company should not be negative
invalid_years = (employee["YearsAtCompany"] < 0).sum()
print(f"Invalid Years at Company (< 0): {invalid_years}")
```

## **7. Save Cleaned Data**

```python
employee.to_csv("employee.csv", index=False)
education.to_csv("education_level.csv", index=False)
performance.to_csv("performance_rating.csv", index=False)
rating.to_csv("rating_level.csv", index=False)
satisfaction.to_csv("satisfied_level.csv", index=False)
```
